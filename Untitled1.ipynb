```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, a, b, c):
    temperature = a * (time**2) + b * time + c
    return temperature, a, b, c

def get_coefficients_from_user():
    a = float(input("Enter the coefficient 'a': "))
    b = float(input("Enter the coefficient 'b': "))
    c = float(input("Enter the coefficient 'c': "))
    return a, b, c

def plot_graph(time_values, temperature_values, label, linestyle='-'):
    plt.plot(time_values, temperature_values, label=label, linestyle=linestyle)

def main():
    # Hardcoded sets of coefficients
    hardcoded_coefficients = {'a': 0.1, 'b': -1, 'c': 30}

    # Get coefficients from the user
    user_coefficients = get_coefficients_from_user()

    # Generate time values from 0 to 10 with 100 points
    time_values = np.linspace(0, 10, 100)

    # Plot with hardcoded coefficients
    temperature_hardcoded, _, _, _ = quadratic_model(time_values, **hardcoded_coefficients)
    plot_graph(time_values, temperature_hardcoded, 'Hardcoded Coefficients')

    # Plot with user-input coefficients
    temperature_user_input, _, _, _ = quadratic_model(time_values, *user_coefficients)
    plot_graph(time_values, temperature_user_input, 'User Input Coefficients', linestyle='--')

    plt.xlabel('time')
    plt.ylabel('temperature')
    plt.legend()
    plt.title('Weather Modeling with Quadratic Equation')
    plt.show()

if __name__ == "__main__":
    main()
```
